# Começando a utilizar modelos de ML

```python
df.head()
```

```python
# Para o método k-means, utiliamos apenas variáveis contínuas

cols = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds',
       'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'Days_customer']
```

```python
df_cluster = df[cols]
df_cluster
```

$$\frac{x - mean}{standard}$$

```python
# A primeira lib se refere ao algoritmo de k-means. A segunda se refere a normalização dos dados
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# Normalizando dados e plotando em boxplot
df_cluster_scaler = pd.DataFrame(scaler.fit_transform(df_cluster), columns=cols)
df_cluster_scaler.plot(kind='box', figsize=(14, 3), rot=45);
```

```python
kmeans = KMeans(4)
```

```python
kmeans.fit_predict(df_cluster_scaler)
```

```python
kmeans.inertia_
```

```python
from tqdm import tqdm
```

```python
ks_lst = range(2, 21)
inertias_lst = []

for i in tqdm(ks_lst):
    kmeans = KMeans(i)
    kmeans.fit(df_cluster_scaler)
    inertias_lst.append(kmeans.inertia_)
```

```python
# O número de clusters ideal é aquele no qual o camminho médio quadrático cai mais bruscamente e se torna contínuo
plt.figure(figsize=(14, 3))
plt.xticks(ks_lst)
plt.plot(ks_lst, inertias_lst);
```

```python
kmeans = KMeans(3)
```

```python
clusters = kmeans.fit_predict(df_cluster_scaler)
clusters
```

```python
df['Clusters'] = clusters
```

```python
df.head()
```

```python
sns.boxplot(data=df, x='Clusters', y='Year_Birth');
```

```python
# Gerando uma função de distribuição de probabilidades

sns.displot(data=df, x='Year_Birth', hue='Clusters', kind='kde');
```

```python
sns.displot(data=df, x='Income', hue='Clusters', kind='kde');
```
